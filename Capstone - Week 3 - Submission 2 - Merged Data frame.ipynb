```python
# importing the neccesary library
import numpy as np
import pandas as pd
```

#### Importing the data from the URL link provided

```python
url = "http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
my_data = pd.read_html(url)[0]
my_data
```

#### Dropping the rows for which there is no value assigned for Borough

```python
my_data.drop(my_data[my_data.Borough == 'Not assigned'].index, inplace=True)
```

```python
my_data.head()
```

```python
# printing the shape of the data frame
my_data.shaped
```

```python
latlong_data = pd.read_csv('Geospatial_Coordinates.csv')
```

```python
latlong_data.head()
```

```python
merged_data = my_data.merge(latlong_data,on=['Postal Code'])
```

```python
merged_data.head(20)
```

```python
# create map of Toronto using latitude and longitude values
import folium
from geopy.geocoders import Nominatim
address = 'Toronto, CA'

geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(merged_data['Latitude'], merged_data['Longitude'], 
                                           merged_data['Borough'], merged_data['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
