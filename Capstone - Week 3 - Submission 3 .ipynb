#### Importing the data from the URL link provided

```python
url = "http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
my_data = pd.read_html(url)[0]
my_data
```

#### Dropping the rows for which there is no value assigned for Borough

```python
my_data.drop(my_data[my_data.Borough == 'Not assigned'].index, inplace=True)
```

```python
latlong_data = pd.read_csv('Geospatial_Coordinates.csv')
```

```python
latlong_data.head()
```

```python
merged_data = my_data.merge(latlong_data,on=['Postal Code'])
```

```python
merged_data.head(20)
```

##### Creating the map using latitude and longitude data from the dataframe and Folium library

```python
# Using the latitude and longitude values from the dartaframe above we create a map of Toronto
import folium
locations = merged_data[['Latitude','Longitude']]
location_list = locations.values.tolist()
map = folium.Map(location= [43.657,-79.378],tiles= 'Stamen Terrain', zoom_start=11)
for i in range(0,len(location_list)):
    folium.Marker(location_list[i],popup=merged_data['Neighbourhood'][i]).add_to(map)
map
```

##### We will now add a marker cluster which will represent the total number of locations which are near each other and will provide a better view of the map

```python
from folium.plugins import MarkerCluster
map_updated = folium.Map(location=[43.657,-79.378],tiles= 'Stamen Terrain', zoom_start=11)
# Each number represents the number of localities which are close together. 
# Clicking on a cluster expands the locations in the cluster
map_cluster = MarkerCluster().add_to(map_updated) 
for i in range(0,len(location_list)):
    folium.Marker(location_list[i],popup=merged_data['Neighbourhood'][i]).add_to(map_cluster)
map_updated
```
